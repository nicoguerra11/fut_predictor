# EDA — Dataset de Mundiales (2014-2026)

Análisis exploratorio del dataset procesado por `data/fetch_data.py`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

plt.style.use('dark_background')
plt.rcParams['figure.dpi'] = 120

df = pd.read_csv('../data/processed/matches.csv')
print(f'Partidos totales: {len(df)}')
df.head()
```

```python
# Distribución de goles por equipo y torneo
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for ax, col, title in zip(axes, ['goles_local', 'goles_visitante'], ['Goles Local', 'Goles Visitante']):
    df[col].value_counts().sort_index().plot(kind='bar', ax=ax, color='#3b82f6', edgecolor='none')
    ax.set_title(title)
    ax.set_xlabel('Goles')
    ax.set_ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
# Media de goles por fase
by_phase = df.groupby('fase')[['goles_local', 'goles_visitante']].mean().round(2)
print('Media de goles por fase:')
print(by_phase)
```

```python
# Ventaja de local
groups_df = df[df['fase'] == 'grupos']
p_home = (groups_df['goles_local'] > groups_df['goles_visitante']).mean()
p_draw = (groups_df['goles_local'] == groups_df['goles_visitante']).mean()
p_away = (groups_df['goles_local'] < groups_df['goles_visitante']).mean()
print(f'Grupos — Local: {p_home:.1%}  Empate: {p_draw:.1%}  Visitante: {p_away:.1%}')
```

```python
# Top 10 goleadores (como local)
df.groupby('equipo_local')['goles_local'].sum().nlargest(10).plot(
    kind='barh', figsize=(9, 4), color='#10b981', edgecolor='none'
)
plt.title('Top 10 equipos — goles como local')
plt.xlabel('Goles totales')
plt.tight_layout()
plt.show()
```

```python
# Relación ranking FIFA vs goles marcados
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df['ranking_local'], df['goles_local'], alpha=0.3, s=20, color='#3b82f6', label='Local')
ax.scatter(df['ranking_visitante'], df['goles_visitante'], alpha=0.3, s=20, color='#10b981', label='Visitante')
ax.set_xlabel('Ranking FIFA (menor = mejor)')
ax.set_ylabel('Goles marcados')
ax.set_title('Ranking FIFA vs goles marcados')
ax.legend()
plt.tight_layout()
plt.show()
```
